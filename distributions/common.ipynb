Common distributions.

- Bernoulli
- Binomial

```python
# common imports
import polars as pl
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# generate some binomial data

n = 1000
p = 0.5
data = np.random.binomial(n, p, 1000)

# draw using seaborn
sns.histplot(data, kde=True)
plt.title(f"Binomial, n = {n}, p = {p}")
plt.show()
```

```python
import scipy.stats as stats

n = 10000
p = 0.3
rv = stats.binom(n, p)
mean, var = rv.stats(moments='mv')
print(mean, var)
```

```python
# poisson variable, can be used to model the number of events in a fixed interval of time or space
data = np.random.poisson(50000, 100000)

# draw using seaborn
sns.histplot(data, kde=True)
plt.title(f"Poisson, lambda = 50000, n = 100000")
plt.show()
```

```python
# Geometry distribution
p = 0.01
data = np.random.geometric(p, 10000)

# draw using seaborn
sns.histplot(data, kde=True)
plt.title(f"Geometric, p = {p}")
plt.show()
```
