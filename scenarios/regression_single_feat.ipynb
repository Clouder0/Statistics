Sometimes we may want to check the effect of one exact variable in multivariate regression.

This notebook showcases how to do it.


```python
"""Use polars dataframe as source."""

import polars as pl
import numpy as np
from sklearn import linear_model, compose, preprocessing, metrics
import matplotlib.pyplot as plt
import seaborn as sns


x_cols = ["X0", "X1"]
y_col = "Y"


def gen_data(seed=0):
    # generate random data, Y = f(X) + noise

    # first determine the distribution of X, choose from Gaussian, Poisson, Uniform

    # init seed
    np.random.seed(seed)

    # random sample size
    sample_size = np.random.randint(100, 1000)
    x_dists = ["Gaussian", "Poisson", "Uniform"]
    res = pl.LazyFrame()
    for col in x_cols:
        x_dist = np.random.choice(x_dists)
        print(x_dist)

        if x_dist == "Gaussian":
            # generate random gaussian with random parameters
            loc = (np.random.random() - 0.5) * 5
            scale = np.random.random() * 10
            x = np.random.normal(loc, scale, sample_size)
        elif x_dist == "Poisson":
            # generate random lambda
            lamb = np.random.random() * 10
            x = np.random.poisson(lamb, sample_size)
        else:
            x = np.random.uniform(0, 1, sample_size)

        res = res.with_columns(pl.Series(x).alias(col))

        # now random choose the relation between Y and x
        y_dist = np.random.choice(["Linear", "Quadratic"])
        if y_dist == "Linear":
            # generate random slope
            slope = np.random.random() * 10
            # generate random intercept
            intercept = np.random.random() * 10
            y = slope * x + intercept
        else:
            # generate random quadratic
            a = (np.random.random() - 0.5) * 2
            b = (np.random.random() - 0.5) * 2
            c = (np.random.random() - 0.5) * 2
            y = a * x**2 + b * x + c

        # y add noise
        y_std = y.std()
        noised_y = y + np.random.normal(0, 0.2 * y_std, sample_size)
        res = (
            res.with_columns(
                pl.Series(y).alias(f"{y_col}_{col}"),
                pl.Series(noised_y).alias(f"{y_col}_{col}_noised"),
            )
            .lazy()
            .collect()
            .lazy()
        )

    # sum up all Y and add noise
    res = res.with_columns(
        (sum([pl.col(f"{y_col}_{col}_noised") for col in x_cols])).alias(y_col)
    )
    res = res.collect()
    # now calculate the variance of Y
    y_std = res[y_col].std()
    # add noice accordingly
    res = res.with_columns(
        (pl.col(y_col) + np.random.normal(0, y_std * 0.2, sample_size)).alias(
            f"{y_col}_noised"
        )
    )
    return res


def plot_distribution(data: pl.DataFrame, col: str):
    # draw hist using seaborn
    sns.histplot(data[col], kde=True)
    plt.title(f"{col}")
    plt.show()


def plot_feature(data: pl.DataFrame, x_col: str, nowy_col: str):
    # use seaborn to draw regplot between x_col and y_col
    # set title, xlabel and ylabel and legend
    sns.regplot(x=x_col, y=nowy_col, data=data, scatter=True)
    plt.title(f"{nowy_col} vs. {x_col}")
    plt.xlabel(x_col)
    plt.ylabel(nowy_col)
    # calculate correlation between x_col and y_col
    corr = np.corrcoef(data[x_col], data[nowy_col])[0, 1]
    # calculate regression and r2
    m = linear_model.LinearRegression()
    m.fit(data[x_col].to_numpy().reshape(-1, 1), data[nowy_col])
    m.predict(data[x_col].to_numpy().reshape(-1, 1))
    r2 = m.score(data[x_col].to_numpy().reshape(-1, 1), data[nowy_col])
    plt.legend([f"corr = {corr}\nr2 = {r2}"])
    plt.show()


class MyModel:
    trans: compose.ColumnTransformer
    # the underlying model should be linear regression
    m: compose.TransformedTargetRegressor

    def __init__(
        self,
        trans: compose.ColumnTransformer,
        m: compose.TransformedTargetRegressor,
        response: str,
    ):
        self.trans = trans
        self.m = m
        self.response = response

    def fit(self, src: pl.DataFrame):
        X = self.trans.fit_transform(src)
        Y = src[self.response]
        self.m.fit(X, Y)

    def predict(self, src: pl.DataFrame):
        X = self.trans.transform(src)
        return src.with_columns(PredY=pl.Series(self.m.predict(X)))


def draw_feat_marginal(data: pl.DataFrame, col: str, model: MyModel):
    # calc PredY - (X except col) * coefficient
    find_idx = -1
    for i, feat in enumerate(model.trans.get_feature_names_out()):
        if str(feat).endswith(col):
            find_idx = i
            break
    if find_idx == -1:
        raise ValueError(f"Cannot find {col} in {model.trans.get_feature_names_out()}")
    print(find_idx)
    actual_Y = data["Y_noised"]
    X = model.trans.transform(data)
    now_X = X[:, find_idx]
    predY = data["PredY"]
    contribute = now_X * model.m.coef_[find_idx]
    data = data.with_columns(
        pl.Series(actual_Y - predY + contribute).alias(f"PredY_no{col}"),
        pl.Series(contribute).alias(f"Contribute_{col}"),
    )
    display(data)
    plot_feature(data, col, f"PredY_no{col}")
    plot_feature(data, col, f"Contribute_{col}")


my_trans = compose.make_column_transformer(
    (preprocessing.StandardScaler(), ["X0", "X1"]),
)
my_reg = linear_model.LinearRegression()
my_model = MyModel(my_trans, my_reg, f"{y_col}_noised")

data = gen_data(8142)
my_model.fit(data)
res = my_model.predict(data)
for col in x_cols:
    plot_distribution(data, col)
    plot_feature(data, col, f"{y_col}_{col}_noised")
    plot_feature(data, col, f"{y_col}_noised")


# now draw predict Y and real Y
r2 = metrics.r2_score(res[y_col], res["PredY"])
print(r2)
plot_feature(res, "PredY", y_col)

print(my_model.m.coef_)
display(res)
draw_feat_marginal(res, "X1", my_model)
```
