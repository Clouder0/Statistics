First we should read `college.csv` using pandas.

```python
import pandas as pd

college = pd.read_csv("college.csv")
college.rename(columns={"Unnamed: 0": "College"}, inplace=True)
college.set_index("College", inplace=True)
college
```

Use describe to observe the overall statistics for our data.

```python
college.describe()
```

We can see that there are 777 colleges. 

Interestingly, there is an at least 1% top students even in the worst college. That's out of my expectation.

Why Grad.Rate can exceed 100???

Use scatter_matrix to see relationship among top10perc,apps,enroll.

```python
pd.plotting.scatter_matrix(college[["Top10perc", "Apps", "Enroll"]]);
```

We can see that, generally enrolls are more at schools with higher "elite rate". And obviously, more acceptance means more enrollment.

Now we will use boxplot to see the relationship between Outstate and Private.

```python
from matplotlib import pyplot as plt

fig, ax = plt.subplots()
college.boxplot(column="Outstate", by="Private", ax=ax);
```

And of course, private schools get less state tuition, so outstate is larger.

Next, we will add a qualitative variable "Elite" to indicate if a college should be considered "Elite College".

```python
college["Elite"] = pd.cut(college["Top10perc"], [0,50,100], labels=["NO", "YES"])
college.value_counts("Elite")
```

There are about 78 elite colleges in total. Now let's see there outstate.

```python
college.boxplot(column="Outstate", by="Elite");
```

And I guess: most elite colleges are private. Let's see if it's true.

```python
import seaborn as sns
import scipy
import scipy.stats

sns.countplot(data=college, x="Private", hue="Elite",stat="percent")

# get correlation between private and elite
df = pd.crosstab(college["Private"], college["Elite"])
print(df)
res = scipy.stats.chi2_contingency(df)
res
```

Seems there exists some correlation between private and elite. Obviously.
